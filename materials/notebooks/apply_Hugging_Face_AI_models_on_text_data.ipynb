```python
import pandas as pd

df = pd.read_csv("hf://datasets/DSCI511G1/COP26_Energy_Transition_Tweets/energy_transition_tweets_final.csv")
df.head()
```

# emotion classification

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

emotion = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
```

# test the model

```python
emotion("I love NLP")
```

```python
sample_df = df.sample(200)
```

```python
# prompt: write code to get the emotion and the score of each tweet in the sample_df

import pandas as pd

# Assuming df and emotion are already defined as in the previous code

# Function to get emotion and score
def get_emotion(text):
    try:
        result = emotion(text)[0]
        return result['label'], result['score']
    except Exception as e:
        print(f"Error processing text: {text}, Error: {e}")
        return None, None


# Apply the function to the sample_df
sample_df[['emotion', 'score']] = sample_df['tweetContent'].apply(lambda x: pd.Series(get_emotion(x)))
```

```python
# Display the updated DataFrame
sample_df[['tweetContent', 'emotion', 'score']]
```

```python
# Save the DataFrame (replace 'sample_df.csv' with your desired file name)
sample_df.to_csv('sample_emotions.csv', index=False)
```

NER


```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-large-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-large-NER")

ner = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My name is Wolfgang and I live in Berlin"

ner_results = ner(example)
print(ner_results)
```

```python
# prompt: write code to extract people, locations and organization from tweetContent in the sample_df using ner model. add people, locations and organization as new columns

import pandas as pd
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForTokenClassification

# ... (previous code)

def extract_entities(text):
    try:
        results = ner(text)
        people = ' '.join([result['word'] for result in results if 'PER' in result['entity'] ])
        locations = ' '.join([result['word'] for result in results if 'LOC' in result['entity'] ])
        organizations = ' '.join([result['word'] for result in results if 'ORG' in result['entity']])
        return people, locations, organizations
    except Exception as e:
        print(f"Error processing text: {text}, Error: {e}")
        return [], [], []

# Apply the function to the sample_df
sample_df[['people', 'locations', 'organizations']] = sample_df['tweetContent'].apply(lambda x: pd.Series(extract_entities(x)))

# Display the updated DataFrame
sample_df[['tweetContent', 'people', 'locations', 'organizations']]
```
